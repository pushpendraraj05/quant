```python
# Date vs Close

from matplotlib import pyplot as plt
import seaborn as sns
def _plot_series(series, series_name, series_index=0):
  palette = list(sns.palettes.mpl_palette('Dark2'))
  xs = series['Date']
  ys = series['Close']

  plt.plot(xs, ys, label=series_name, color=palette[series_index % len(palette)])

fig, ax = plt.subplots(figsize=(10, 5.2), layout='constrained')
df_sorted = cof.sort_values('Date', ascending=True)
_plot_series(df_sorted, '')
sns.despine(fig=fig, ax=ax)
plt.xlabel('Date')
_ = plt.ylabel('Close')
```

## **Correaltion Heatmap**

```python

# Concatenate the two dataframes horizontally
combined_df = pd.concat([cof, hdfc], axis=1)

# Calculate the correlation matrix
correlation_matrix = combined_df.corr()

# Plot the correlation matrix heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", annot_kws={"size": 10})
plt.title('Correlation Matrix Heatmap')
plt.show()
```

### It identify relationships and dependencies between variables.

## CardOne Close values change over time.

```python
start_date = '2023-04-04'
end_date = '2024-04-03'
filtered_df = cof.loc[start_date:end_date]

# Plotting the time series data
plt.figure(figsize=(10, 6))

plt.plot(filtered_df.index, filtered_df['Close'])

plt.xlabel('Date')
plt.ylabel('Close Price')
plt.title('Time Series Data (04-04-2023 to 02-04-2024)')
plt.grid(True)
plt.show()
```

```python
filtered_df = cof.loc[start_date:end_date, 'Close']

# Plotting histograms and density plots
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.hist(filtered_df, bins=20, density=True, color='skyblue', edgecolor='black', alpha=0.7)
plt.title('Histogram of Close Prices')
plt.xlabel('Close Price')
plt.ylabel('Density')

plt.subplot(1, 2, 2)
filtered_df.plot(kind='density', color='green')
plt.title('Density Plot of Close Prices')
plt.xlabel('Close Price')
plt.ylabel('Density')

plt.tight_layout()
plt.show()
```

### It show the distribution of data among the given Closing Price

```python
filtered_df = cof.loc[start_date:end_date, 'Open']

# Plotting histograms and density plots
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.hist(filtered_df, bins=20, density=True, color='skyblue', edgecolor='black', alpha=0.7)
plt.title('Histogram of Close Prices')
plt.xlabel('Open Price')
plt.ylabel('Density')

plt.subplot(1, 2, 2)
filtered_df.plot(kind='density', color='green')
plt.title('Density Plot of Open Prices')
plt.xlabel('Open Price')
plt.ylabel('Density')

plt.tight_layout()
plt.show()
```

### It show the distribution of data among the given Opening Price
